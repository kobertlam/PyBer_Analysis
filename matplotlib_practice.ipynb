```python
# Import dependencies.
#
# Pyplot provides a MATLAB-like interface for making plots
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Method 1: use MATHLAB method
#
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Method 2: use the object-oriented interface method
#
# Create the plot with ax.plt()
#
# To create a figure
# plt.subplots() function returns a tuple that contains a figure and axes objects
# fig and ax are the object references
fig, ax = plt.subplots()
# To define the axes
ax.plot(x_axis, y_axis)
```

```python
# Method 3: use the object-oriented interface method
# Create the plot with ax.plt()
# here "fig" references the figure, and "ax" references the axes of the figure, and allow us to annotate the figure
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Method 4: use the object-oriented interface method
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# (5.1.7) Scatter Plot (MATLAB method)
#
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Scatter Plot: Method 1
# Using plt.plot() and add 'o' as a parameter to switch the plot from a line chart to a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Scatter Plot: Method 2
plt.scatter(x_axis, y_axis)
```

```python
# Annotate the scatter plot
plt.scatter(y_axis, x_axis, c="r", label="Chicago")
# Add title and axes label
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Invert the y-axis data
plt.gca().invert_yaxis()
# Add the legend.
plt.legend()
```

```python
# Create a Bubble Chart using MATLAB method
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjust the size of the bubble by multiplying the data in the y-axis by any value
# Let's multiply each data point in the y-axis by 3
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
# use the new y-axis list for the s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Annotate the Bubble Chart using the Object-Oriented Interface
# increase the size by a factor of 5 and add a black edge color to the circles that have a line width of 2
# color = skyblue, with 20% transparent
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis,s=[i * 5 for i in y_axis],
           c="skyblue", alpha=0.8, linewidths=2, edgecolors="black",
           label="Boston")
# Label x & y-axis
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Set the y limit between 0 and 45.
ax.set_xlim(0, 45)
# Invert the y-axis data
ax.invert_yaxis()
# Add title
ax.set_title("PyBer Fare by Month")
# Add a legend
ax.legend()
```

```python
# (5.1.8) Create a Pie Chart using MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
# increase the size of the pie chart
plt.subplots(figsize=(8, 8))
# Explode will offset the indicated wedge by a fraction of the radius:
# 0 = zero distance, 1 = completely outside the diameter of the pie
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
